```python
# 1) Função para calcular IMC e devolver um dicionário
def calcula_imc_dict(peso=70, altura=1.75):
    imc = peso / (altura * altura)
    # montamos um dicionário com peso, altura e IMC
    resultado = {
        "peso": peso,
        "altura": altura,
        "imc": imc
    }
    return resultado

import statistics
import random

def main():
    # Teste do cálculo de IMC em dicionário
    imc_info = calcula_imc_dict(80, 1.80)
    print("Dados do IMC:", imc_info)

    # Mesma lista de alturas
    alturas = [45, 23, 67, 12, 11, 89, 23, 41, 50, 62,
               78, 34, 56, 19, 72, 88, 11, 90, 39, 65,
               76, 27, 48, 59, 81, 14, 11, 93, 3, 68,
               29, 52, 74, 16, 85, 20, 55, 38, 69, 11,
               83, 7, 44, 61, 18, 96, 22, 58, 31, 71,
               40, 53, 87, 31]

    # Estatísticas guardadas num dicionário
    stats = {}
    stats["maior"] = max(alturas)
    stats["menor"] = min(alturas)
    stats["soma"] = sum(alturas)
    stats["abs_-25"] = abs(-25)
    stats["round_3.14159"] = round(3.14159, 2)
    stats["media"] = statistics.mean(alturas)
    stats["mediana"] = statistics.median(alturas)
    stats["moda"] = statistics.mode(alturas)
    stats["variancia"] = statistics.variance(alturas)
    stats["desvio_padrao"] = statistics.stdev(alturas)

    print("\nEstatísticas das alturas (dicionário):")
    for chave, valor in stats.items():
        print(f"{chave} -> {valor}")

    # Matrizes dentro de um dicionário
    matrizes = {}

    # floats 5x10
    matrizes["floats_5x10"] = []
    for i in range(5):
        linha = []
        for j in range(10):
            linha.append(random.random())
        matrizes["floats_5x10"].append(linha)

    # ints 8x4
    matrizes["ints_8x4"] = []
    for i in range(8):
        linha = []
        for j in range(4):
            linha.append(random.randint(0, 9))
        matrizes["ints_8x4"].append(linha)

    print("\nMatrizes armazenadas em dicionário:")
    print("floats_5x10:")
    for linha in matrizes["floats_5x10"]:
        print(linha)
    print("\nints_8x4:")
    for linha in matrizes["ints_8x4"]:
        print(linha)

if __name__ == "__main__":
    main()
```
