```python
# 1) Função para calcular IMC
def calcula_imc(peso=70, altura=1.75):
    # IMC = peso dividido pela altura ao quadrado
    imc = peso / (altura * altura)
    return imc

# 2) Trabalhando com a lista de alturas
import statistics   # só para média, mediana, moda, etc.
import random       # para gerar números aleatórios

def main():
    # Testando a função de IMC
    print("IMC padrão (70 kg, 1.75 m):", calcula_imc())
    print("IMC para 80 kg e 1.80 m:", calcula_imc(80, 1.80))

    # Lista de alturas (valores quaisquer)
    alturas = [45, 23, 67, 12, 11, 89, 23, 41, 50, 62,
               78, 34, 56, 19, 72, 88, 11, 90, 39, 65,
               76, 27, 48, 59, 81, 14, 11, 93, 3, 68,
               29, 52, 74, 16, 85, 20, 55, 38, 69, 11,
               83, 7, 44, 61, 18, 96, 22, 58, 31, 71,
               40, 53, 87, 31]

    # Funções padrão
    maior = max(alturas)
    menor = min(alturas)
    soma = sum(alturas)
    valor_absoluto = abs(-25)
    arredondado = round(3.14159, 2)

    print("\n-- Funções básicas --")
    print("Maior valor:", maior)
    print("Menor valor:", menor)
    print("Soma dos valores:", soma)
    print("Valor absoluto de -25:", valor_absoluto)
    print("3.14159 arredondado (2 casas):", arredondado)

    # Funções do módulo statistics
    media = statistics.mean(alturas)
    mediana = statistics.median(alturas)
    moda = statistics.mode(alturas)
    variancia = statistics.variance(alturas)
    desvio = statistics.stdev(alturas)

    print("\n-- Estatísticas --")
    print("Média:", media)
    print("Mediana:", mediana)
    print("Moda:", moda)
    print("Variância amostral:", variancia)
    print("Desvio padrão amostral:", desvio)

    # 3) Matrizes aleatórias
    # 3.1) Matriz 5x10 de floats entre 0 e 1
    matriz_floats = []
    for i in range(5):
        linha = []
        for j in range(10):
            n = random.random()  # valor entre 0.0 e 1.0
            linha.append(n)
        matriz_floats.append(linha)

    print("\nMatriz 5×10 de floats:")
    for linha in matriz_floats:
        print(linha)

    # 3.2) Matriz 8x4 de inteiros entre 0 e 9
    matriz_ints = []
    for i in range(8):
        linha = []
        for j in range(4):
            n = random.randint(0, 9)  # inteiro de 0 a 9
            linha.append(n)
        matriz_ints.append(linha)

    print("\nMatriz 8×4 de inteiros:")
    for linha in matriz_ints:
        print(linha)

if __name__ == "__main__":
    main()
```
